1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

    Answer: In hypothesis testing, Type I and Type II errors are two kinds of mistakes that can happen when we make decisions about a null hypothesis (H₀).
    
    A Type I error happens when we reject the null hypothesis even though it is actually true. In simple words, we think something is happening when it really isn’t. This is called a false positive. The probability of making this mistake is represented by α (alpha), usually set at 0.05. For example, if we say a new medicine works when it actually doesn’t, that is a Type I error.
    
    A Type II error, on the other hand, occurs when we fail to reject the null hypothesis even though it is false. This means we miss something that is actually there, also known as a false negative. The probability of this error is represented by β (beta). For example, if a medicine actually works but we conclude that it doesn’t, it is a Type II error. The power of a test (1 − β) tells us how good the test is at detecting a real effect.
    
    These errors play an important role in decision-making because reducing one often increases the other. For instance, if we try to be very strict and reduce Type I errors, we might increase the chances of Type II errors. The importance of each error depends on the situation. In medical testing, missing a real disease (Type II error) can be dangerous, so it should be minimized. In other areas like quality control, allowing defective products (also a Type II error) can cause serious problems.

2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

    Answer: The p-value in hypothesis testing is the probability of obtaining results that are as extreme as (or more extreme than) the observed data, assuming that the null hypothesis (H₀) is true. The p-value helps us understand how likely our observed result is if there is actually no real effect or difference.

    Interpretation:
    
    If the p-value is small (≤ α, e.g., 0.05) → the observed result is unlikely under H₀, so we reject the null hypothesis. This suggests that there is statistical evidence in favor of the alternative hypothesis.
    
    If the p-value is large (> α) → the observed result is consistent with H₀, so we fail to reject the null hypothesis. This means there is not enough evidence to support the alternative hypothesis.

3. Explain the difference between a Z-test and a T-test, including when to use
each.

    Answer: A Z-test and a T-test are both statistical methods used in hypothesis testing to determine whether there is a significant difference between sample data and a population parameter. The main difference between them lies in the information available about the population and the sample size.

    A Z-test is used when the population standard deviation is known and the sample size is large (𝑛≥30n≥30). It assumes that the data follows a normal distribution or that the sample size is large enough for the Central Limit Theorem to apply. Because σ is known, the test statistic follows a standard normal distribution (Z-distribution).
    For example, if we know the population standard deviation of heights and want to test whether a sample mean differs from the population mean, we use a Z-test.

    In contrast, a T-test is used when the population standard deviation is unknown and must be estimated using the sample standard deviation. It is especially suitable for small sample sizes (𝑛<30n). The test statistic follows a t-distribution, which is similar to the normal distribution but has heavier tails to account for extra uncertainty.For example, if we want to test the average marks of a small class where the population standard deviation is not known, we use a T-test.

4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

    Answer: A confidence interval is a range of values used to estimate an unknown population parameter (such as a mean or proportion) based on sample data. Instead of giving a single value, it provides an interval within which the true population value is expected to lie with a certain level of confidence.

    A confidence interval is typically written as:

    Estimate ± Margin of Error

    The margin of error represents the uncertainty in the estimate and plays a key role in determining the width of the confidence interval. A larger margin of error results in a wider interval, indicating less precision but greater confidence that the true value lies within the range. On the other hand, a smaller margin of error produces a narrower interval, which is more precise but may reflect lower certainty.

    In terms of interpretation, a confidence interval provides a range of plausible values for the population parameter rather than a definite answer. The margin of error helps us understand how reliable and precise this estimate is.


5. Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

    Answer: The ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are significant differences between the means of three or more groups.

    The purpose of ANOVA is to check whether the variation between group means is large enough, compared to the variation within the groups, to conclude that at least one group mean is different. It does this by analyzing two types of variability:

    1. Between-group variation (differences among group means)

    2. Within-group variation (variability within each group)

    The hypotheses in ANOVA are:

    -> Null hypothesis (H₀): All group means are equal

    -> Alternative hypothesis (H₁): At least one group mean is different

    Assumptions of ANOVA:

    1. Independence of observations:

        Each sample is independent of the others.

    2. Normality:

        The data in each group should be approximately normally distributed.

    3. Homogeneity of variance:

        The variances of all groups should be equal.

    ANOVA hypothesis testing:

    A t-test is limited to comparing two groups only, but ANOVA extends this concept to three or more groups by comparing all group means at once using an F-statistic. This avoids performing multiple pairwise tests, which would increase the probability of making a Type I error.

    If the ANOVA result is significant, it tells us that at least one group is different, but it does not specify which one. To identify the specific groups, we perform Tukey's test.

6.  Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from scipy import stats

# Sample data
data = [62, 65, 60, 63, 64, 62, 61, 66, 63, 62]

# Known population mean and standard deviation
population_mean = 50
population_std = 5

# Sample size
n = len(data)

# Sample mean
sample_mean = np.mean(data)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

7. Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate population (non-normal distribution)
population = np.random.exponential(scale=2, size=10000)

# Step 2: Function to compute sample means
def sample_means(sample_size, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        means.append(np.mean(sample))
    return means

# Step 3: Generate sample means for different sample sizes
means_5 = sample_means(5)
means_30 = sample_means(30)
means_50 = sample_means(50)

# Step 4: Plot histograms
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(means_5, bins=30, edgecolor='black')
plt.title("Sample Size = 5")

plt.subplot(1, 3, 2)
plt.hist(means_30, bins=30, edgecolor='black')
plt.title("Sample Size = 30")

plt.subplot(1, 3, 3)
plt.hist(means_50, bins=30, edgecolor='black')
plt.title("Sample Size = 50")

plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval_visual(data, confidence=0.95):
    # Convert to numpy array
    data = np.array(data)

    # Sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error

    # Confidence interval
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Interval ({confidence*100}%): ({lower:.2f}, {upper:.2f})")

    # Visualization
    plt.hist(data, bins=10, edgecolor='black', alpha=0.7)

    # Mean line
    plt.axvline(mean, linestyle='--', label='Mean')

    # Confidence interval lines
    plt.axvline(lower, linestyle=':', label='Lower CI')
    plt.axvline(upper, linestyle=':', label='Upper CI')

    plt.title("Confidence Interval Visualization")
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

# Example usage
data = [48, 52, 50, 49, 51, 53, 47, 52, 50, 49]
confidence_interval_visual(data)
```

10. Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([25, 30, 20, 25])

# Expected frequencies (example: equal distribution)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

The given Python code performs a Chi-square goodness-of-fit test by comparing the observed frequencies [25, 30, 20, 25] with the expected frequencies [25, 25, 25, 25]. After running the code, a p-value is obtained. If the p-value is greater than 0.05, it indicates that the difference between observed and expected values is small and likely due to random variation. In this case, we fail to reject the null hypothesis and conclude that the observed distribution is consistent with the expected distribution.

If the p-value is less than 0.05, it means the difference is statistically significant. In such a case, we reject the null hypothesis and conclude that the observed data does not follow the expected distribution.
